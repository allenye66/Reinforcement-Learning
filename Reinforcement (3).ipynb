```python
# %% [code]
import gym
import random
import time
import numpy as np
from IPython.display import clear_output
env = gym.make("FrozenLake-v0").env
env.render()

# %% [code]
action_space_size = env.action_space.n
state_space_size = env.observation_space.n
print(action_space_size)
print(state_space_size)

# %% [code]
q_table = np.zeros((state_space_size, action_space_size)) #16 by 4
print(q_table)

# %% [code]
epochs = 1000
max_steps = 200

learning_rate = 0.1
discount_rate = 0.99

exploration_rate = 1
max_exploration_rate = 1
min_exploration_rate = 0.01
exploration_decay_rate = 0.01

# %% [code]
rewards_all = []

for episode in range(epochs):
    state = env.reset()
    done = False
    current_reward = 0
    for step in range(max_steps):
        exploration_rate_threshold = random.uniform(0,1)
        if exploration_rate_threshold > exploration_rate:
            action = np.argmax(q_table[state,:])
        else:
            action = env.action_space.sample()
            
        new_state, reward, done, info = env.step(action)
        
        q_table[state, action] = q_table[state, action] * (1 - learning_rate) + learning_rate*(reward  + discount_rate * np.max(q_table[new_state,:])) 
        
        state = new_state
        current_reward += reward
        
        if done == True:
            break
        
    exploration_rate = min_exploration_rate + (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate*episode)
    
    rewards_all.append(current_reward)
    
rewards_print = np.split(np.array(rewards_all), epochs/1000)
#env.render()

accuracy = []
for r in rewards_print:
    accuracy.append(sum(r/1000))

# %% [code]
count = 1000
print("*******Average accuracy per thousand episodes******\n")
for i in accuracy:
    print(count, ": ", i)
    count += 1000

# %% [code]
print("max:", np.amax(accuracy), "at: ", 1+np.argmax(accuracy, axis=0))

# %% [code]
```
